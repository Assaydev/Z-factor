```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for sample and control signals assuming a normal distribution
np.random.seed(0)

# Sample and control data for different Z-factor scenarios
sample_data_z1 = np.random.normal(loc=108, scale=5, size=1000)  # Z-factor close to 0.8
control_data_z1 = np.random.normal(loc=8, scale=2, size=1000)

sample_data_z2 = np.random.normal(loc=38, scale=3, size=1000)  # Z-factor close to 0.5 
control_data_z2 = np.random.normal(loc=8, scale=2, size=1000)

sample_data_z3 = np.random.normal(loc=28, scale=4, size=1000)  # Z-factor close to 0.1
control_data_z3 = np.random.normal(loc=8, scale=2, size=1000)

sample_data_z4 = np.random.normal(loc=32, scale=5, size=1000)  # Z-factor close to 0 
control_data_z4 = np.random.normal(loc=8, scale=3, size=1000)

sample_data_z5 = np.random.normal(loc=29, scale=6, size=1000)  # Z-factor close to -1 
control_data_z5 = np.random.normal(loc=16, scale=3, size=1000)

# Plotting separate histograms for different Z-factor scenarios side by side
fig, axs = plt.subplots(5, 1, figsize=(10, 10))

# Z-factor close to 1 (excellent)
axs[0].hist(sample_data_z1, bins=30, alpha=0.5, label='Sample', color='blue')
axs[0].hist(control_data_z1, bins=30, alpha=0.5, label='Control', color='orange')
axs[0].set_title('Z-factor ≈ 0.8')

# Z-factor close to 0.5
axs[1].hist(sample_data_z2, bins=30, alpha=0.5, label='Sample', color='green')
axs[1].hist(control_data_z2, bins=30, alpha=0.5, label='Control', color='red')
axs[1].set_title('Z-factor ≈ 0.5')

# Z-factor close to 0.1
axs[2].hist(sample_data_z3, bins=30, alpha=0.5, label='Sample', color='purple')
axs[2].hist(control_data_z3, bins=30, alpha=0.5, label='Control', color='yellow')
axs[2].set_title('Z-factor ≈ 0.1')

# Z-factor close to 0
axs[3].hist(sample_data_z4, bins=30, alpha=0.5, label='Sample', color='gray')
axs[3].hist(control_data_z4, bins=30, alpha=0.5, label='Control', color='red')
axs[3].set_title('Z-factor ≈ 0')

# Z-factor close to -1
axs[4].hist(sample_data_z5, bins=30, alpha=0.5, label='Sample', color='blue')
axs[4].hist(control_data_z5, bins=30, alpha=0.5, label='Control', color='red')
axs[4].set_title('Z-factor ≈ -1')

for ax in axs.flat:
    ax.set_xlabel('Signal')
    ax.set_ylabel('Frequency')
    ax.legend()

plt.tight_layout()
plt.savefig('z-factor.png',facecolor="white", transparent=True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for sample and control signals assuming a normal distribution
np.random.seed(0)

# Sample and control data for different Z-factor scenarios
sample_data_z1 = np.random.normal(loc=108, scale=5, size=1000)  # Z-factor close to 1
control_data_z1 = np.random.normal(loc=8, scale=2, size=1000)

sample_data_z2 = np.random.normal(loc=38, scale=3, size=1000)  # Z-factor between 0.5 and 1
control_data_z2 = np.random.normal(loc=8, scale=2, size=1000)

sample_data_z3 = np.random.normal(loc=28, scale=4, size=1000)  # Z-factor between 0 and 0.5
control_data_z3 = np.random.normal(loc=8, scale=2, size=1000)

sample_data_z4 = np.random.normal(loc=29, scale=5, size=1000)  # Z-factor between 0 and 0.5
control_data_z4 = np.random.normal(loc=8, scale=3, size=1000)

sample_data_list = [sample_data_z1, sample_data_z2, sample_data_z3, sample_data_z4]
control_data_list = [control_data_z1, control_data_z2, control_data_z3, control_data_z4]

# Plotting separate histograms for different Z-factor scenarios side by side
fig, axs = plt.subplots(4, 1, figsize=(10, 10))

# Z-factor close to 1 (excellent)
axs[0].hist(sample_data_z1, bins=30, alpha=0.5, label='Sample', color='blue')
axs[0].hist(control_data_z1, bins=30, alpha=0.5, label='Control', color='orange')
axs[0].set_title('Z-factor ≈ 0.8')

# Z-factor between 0.5 and 1 (good)
axs[1].hist(sample_data_z2, bins=30, alpha=0.5, label='Sample', color='green')
axs[1].hist(control_data_z2, bins=30, alpha=0.5, label='Control', color='red')
axs[1].set_title('Z-factor ≈ 0.5')

# Z-factor between 0 and 0.5 (less reliable)
axs[2].hist(sample_data_z3, bins=30, alpha=0.5, label='Sample', color='purple')
axs[2].hist(control_data_z3, bins=30, alpha=0.5, label='Control', color='yellow')
axs[2].set_title('Z-factor ≈ 0.1')

# Z-factor between 0 and 0.5 (less reliable)
axs[3].hist(sample_data_z4, bins=30, alpha=0.5, label='Sample', color='purple')
axs[3].hist(control_data_z4, bins=30, alpha=0.5, label='Control', color='yellow')
axs[3].set_title('Z-factor ≈ 0')

for i in range(4):
    mean_sample = np.mean(sample_data_list[i])
    std_sample = np.std(sample_data_list[i])
    mean_control = np.mean(control_data_list[i])
    std_control = np.std(control_data_list[i])

    # Adding arrows for Z-factor conditions
    axs[i].annotate('Sample Mean', xy=(mean_sample, 50), xytext=(mean_sample + 1, 100),
                       arrowprops=dict(facecolor='black', arrowstyle='->'))
    axs[i].annotate('Control Mean', xy=(mean_control, 50), xytext=(mean_control - 1, 100),
                       arrowprops=dict(facecolor='black', arrowstyle='->'))
    axs[i].annotate('Sample Std Dev', xy=(mean_sample + std_sample, 100), xytext=(mean_sample - std_sample, 100),
                       arrowprops=dict(arrowstyle='<->'))
    axs[i].annotate('Control Std Dev', xy=(mean_control + std_control, 100), xytext=(mean_control - std_control, 100),
                       arrowprops=dict(arrowstyle='<->'))
    

for ax in axs.flat:
    ax.set_xlabel('Signal')
    ax.set_ylabel('Frequency')
    ax.legend()

plt.tight_layout()
plt.show()
```
